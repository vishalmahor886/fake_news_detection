```python
df=df.drop(columns=['title', 'subject', 'date'], axis=1)
```

```python
df.head()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
nltk.download('stopwords')
```

```python
ps = PorterStemmer()
stop_words = set(stopwords.words('english'))
```

```python
def transform_text(text):
    text = text.lower()
    text = nltk.word_tokenize(text)

    y = []

    for i in text:
        if i.isalnum() and i not in stop_words:
            y.append(ps.stem(i))
    return ' '.join(y)
```

```python
df['text'] = df['text'].apply(transform_text)
```

```python
corpus = df['text'].tolist()
```

```python
corpus[0]
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
vec_size=5000
onehot_rep = [one_hot(words, vec_size) for words in corpus]
```

```python
onehot_rep[0]
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, LSTM
```

```python
## embedding representation

sent_length=300
embedded_doc = pad_sequences(onehot_rep, padding='post', maxlen=sent_length)
```

```python
embedded_doc[0]
```

```python
## Create Model 

embedding_vector_features = 50
model=Sequential()
model.add(Embedding(vec_size, 
                    embedding_vector_features,
                    input_length=sent_length))
model.add(LSTM(50))
model.add(Dense(
    1,
    activation='sigmoid'
))
```

```python
model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']

)
```

```python
print(model.summary())
```

```python
X = df.drop(columns=['target'], axis=1)
y = df['target']
```

```python
len(embedded_doc),y.shape
```

```python
X_final = np.array(embedded_doc)
y_final = np.array(y)
```

```python
##train test split

X_train, X_test, y_train, y_test = train_test_split(X_final,y_final,test_size=0.33, random_state=42)
```

```python
## final Tunning

model.fit(
    X_train, 
    y_train,
    validation_data=(
        X_test,
        y_test
    ),
    epochs=10,
    batch_size=128
)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred=np.where(y_pred > 0.6, 1,0) ##AUC ROC Curve
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
